```python
import yfinance
import pandas as pd
from datetime import datetime, timedelta
```

```python
sensex_tickers = [
    "RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "INFY.NS", "HINDUNILVR.NS",
    "ICICIBANK.NS", "HDFC.NS", "BAJFINANCE.NS", "KOTAKBANK.NS", "ITC.NS",
    "SBIN.NS", "BHARTIARTL.NS", "ASIANPAINT.NS", "AXISBANK.NS", "LT.NS",
    "HCLTECH.NS", "MARUTI.NS", "SUNPHARMA.NS", "NESTLEIND.NS", "TITAN.NS",
    "ULTRACEMCO.NS", "ONGC.NS", "TATAMOTORS.NS", "POWERGRID.NS", "NTPC.NS",
    "TATASTEEL.NS", "BAJAJ-AUTO.NS", "MM.NS", "TECHM.NS", "INDUSINDBK.NS"
]
```

```python
start_date = input("Enter start date (YYYY-MM-DD): ")
end_date = input("Enter end date (YYYY-MM-DD): ")
```

```python
def download(ticker):
  print(f'downloading data for {ticker}')
  data = yfinance.download(ticker, start=start_date, end=end_date)
  if not data.empty:
    filename = f"{ticker.replace('.NS','')}_data.csv"
    data.to_csv(filename)
    print(f'data saved to {filename}')

  else:
    print(f'no data available for {ticker}')
```

```python
for ticker in sensex_tickers:
  download(ticker)
```
